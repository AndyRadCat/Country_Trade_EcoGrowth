## Experimenting with the line

```python
exp_BR = countries[['Year','EX_BoP_BR']]
```

```python
exp_BR = exp_BR.assign(t=range(1, len(exp_BR) + 1))
```

```python
exp_BR.head()
```

```python
exp_BR = exp_BR.assign(const=1)
```

```python
exp_BR.head()
```

```python
x = exp_BR[['const', 't']]
y = exp_BR['EX_BoP_BR']

lin_reg = sm.OLS( y, x)
results = lin_reg.fit()
```

```python
results.summary()
```

        Bop(revenue) = -3.961e+10 + 7.55e+09*t

```python
exp_BR.head()
```

```python
fig, ax = plt.subplots(figsize=(10,6))

sns.lineplot(data=exp_BR,
             x="t",
             y="EX_BoP_BR");

sm.graphics.abline_plot(model_results=results, color="green", ax=ax);
```

```python
exp_BR = exp_BR.assign(pred_revenue=results.predict(x))
```

```python
def fit_values(t):
    return -3.961e+10 + (7.55e+09 * t)
```

```python
fit_values(15)
```

```python
exp_BR = exp_BR.assign(manual_pred_revenue=exp_BR['t'].apply(fit_values))
```

```python
exp_BR.head()
```

```python
sns.lineplot(x ='Year', 
            y = 'EX_BoP_BR',
             color='lightcoral',
             data = countries);
```

```python
countries[countries['Year'] == 2018]
```

```python
lb = 2007
ub = 2017
countries_exp = countries[['IM_BoP_BR', 'EX_BoP_BR']]
baseline = countries_exp.loc[countries_exp['IM_BoP_BR'].between(lb, ub)]
```

```python
countries_exp.plot()
```

```python
baseline = (baseline
            .assign(t=range(40, 40 + len(baseline)))
            .assign(const=1))
```

```python
baseline = baseline.assign(pred_revenue=results.predict(baseline[['const', 't']]))
```

```python
exp_BR = pd.concat([exp_BR, baseline])
```

```python
exp_BR.head()
```

```python
mean_BR = exp_BR['EX_BoP_BR'].mean()
```

```python
fig, ax = plt.subplots(figsize=(10,6))

sns.lineplot(data=exp_BR,
            x='Year', 
            y='EX_BoP_BR',
            ax=ax);
sns.lineplot(data=exp_BR,
            x='Year', 
            y='pred_revenue', 
            color='pink');

ax.axhline(mean_BR, color='purple');
```
