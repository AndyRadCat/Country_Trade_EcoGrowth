```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib
import statsmodels.api as sm

%matplotlib inline
sns.set_palette("pastel")
```

## Permorfing statistical analysis

```python
countries = pd.read_csv('df_countries.csv')
```

```python
# dropping the extra columns with years
countries.drop(columns=['Year_MX', 'Year_AR', 'Year_CH', 'Year_ID', 'Year_JP', 'Year_KR', 'Year_IN'])
```

```python
countries = countries.rename(columns={'Year_BR': 'Year'})
```

```python
countries.head()
```

```python
sns.lineplot(x ='Year', 
            y = 'EX_GDP_P_BR',
             color='lightcoral',
             data = countries);
sns.lineplot(x ='Year', 
            y = 'EX_GDP_P_JP', 
             data = countries);
```

```python
sns.lineplot(x ='Year', 
            y = 'EX_GDP_P_AR',
             color='lightcoral',
             data = countries);
sns.lineplot(x ='Year', 
            y = 'EX_GDP_P_IN',
             data = countries);
```

```python
sns.lineplot(x ='Year', 
            y = 'EX_GDP_P_MX',
             color='lightcoral',
             data = countries);
sns.lineplot(x ='Year', 
            y = 'EX_GDP_P_ID',
             data = countries);
```

```python
sns.lineplot(x ='Year', 
            y = 'EX_GDP_P_CH',
             color='pink',
             data = countries);
sns.lineplot(x ='Year', 
            y = 'EX_GDP_P_KR',
             color='purple',
             data = countries);
```

```python
countries.corr()
```

```python
exports_gdp = countries[['EX_GDP_P_BR', 'EX_GDP_P_MX', 'EX_GDP_P_AR', 'EX_GDP_P_CH', 'EX_GDP_P_ID', 'EX_GDP_P_JP', 'EX_GDP_P_KR', 'EX_GDP_P_IN']]
```

```python
exports_gdp.head()
```

```python
heatmap = sns.heatmap(exports_gdp.corr(),
                      vmin=-1,
                      vmax=1, 
                      annot=True)

heatmap.set_title('Exports GDP Correlation', 
                  fontdict={'fontsize':16},
                 pad=12);
```

```python
imports_gdp = countries[['IM_GDP_P_BR', 'IM_GDP_P_MX', 'IM_GDP_P_AR', 'IM_GDP_P_CH', 'IM_GDP_P_ID', 'IM_GDP_P_JP', 'IM_GDP_P_KR', 'IM_GDP_P_IN']]
```

```python
heatmap = sns.heatmap(imports_gdp.corr(),
                      vmin=-1,
                      vmax=1, 
                      annot=True)

heatmap.set_title('Imports GDP Correlation', 
                  fontdict={'fontsize':16},
                 pad=12);
```
